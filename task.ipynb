```python
# get features and labels
data_labels = data['Class']
data_features = data.drop(['Class'], axis=1)
data_features = data[data_features.columns.values]
```

```python
data_features
```

```python
# split the dataset as %30 test and %70 train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_features, data_labels, test_size=0.3)
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier

class Trainer:
    def __init__(self, features, labels) -> None:
        self.features = features
        self.labels = labels
        self.svm = SVC(kernel='linear', gamma='auto', probability=True)
        self.log_reg = LogisticRegression(random_state=0, class_weight="auto")
        self.gaussian = GaussianNB()
        self.knn = KNeighborsClassifier(n_neighbors=2)
        self.randon_forest = RandomForestClassifier()
    
    def train_models(self):
        log_reg_model = self.log_reg.fit(self.features, self.labels)
        print("LogisticRegression model trained")
        gaussian_model = self.gaussian.fit(self.features, self.labels)
        print("GaussianNB model trained")
        knn_model = self.knn.fit(self.features, self.labels)
        print("KNeighborsClassifier model trained")
        random_forest_model = self.randon_forest.fit(self.features, self.labels)
        print("RandomForestClassifier model trained")
        svm_model = self.svm.fit(self.features, self.labels)
        print("SVM model trained")

        return svm_model, log_reg_model, gaussian_model, knn_model, random_forest_model
```

```python
trainer = Trainer(features=X_train, labels=y_train)
models = trainer.train_models()
```

As we know the dataset is unbalanced. Therefore, I have to penalize class 0. The formula for ratio of class_1 = (class_0_count/class_1_count)*class_1_count

```python
import numpy as np
def calculate_accuracy(true:np.array, pred:np.array):
    top_value = data.Class.value_counts().values[0] * 2
    class_1_ratio = data.Class.value_counts().values[0]/data.Class.value_counts().values[1]
    correct_counter = 0
    for idx in range(true.shape[0]):
        if true[idx] == pred[idx]:
            if true[idx] == 1:
                correct_counter += class_1_ratio
            else:
                correct_counter += 1
    return correct_counter / top_value

for model in models:
    y_pred = model.predict(X_test)
    acc = calculate_accuracy(true=y_test.values, pred=y_pred)
    print(f"{str(model)} has {acc} accuracy")
```

As we can see the accuracy is horrible. Our model overfitted and I already expect that overfitting because the dataset unbalanced. Let's try find best threshold for each model with ROC curve

# Goal 2 Optimize your model to reach the best possible performance.

```python
from sklearn.metrics import roc_curve
from numpy import sqrt, argmax

best_thresholds = np.empty((len(models)), dtype=object)
fig, ax = plt.subplots(figsize=(17, 8))
plt.title("ROC curve")
for i, model in enumerate(models):
    model_name = str(model).split("(")[0]
    temp = {
        "model_name": model_name,
        "optimal_thresh": float
    }
    prodiction_prob = model.predict_proba(X_test)
    # keep probabilities for the positive outcome only
    prodiction_prob = prodiction_prob[:, 1]
    fpr, tpr, thresholds = roc_curve(y_test, prodiction_prob)
    # calculate the geometric mean for each threshold
    gmeans = sqrt(tpr * (1-fpr))
    # locate the index of the largest g-mean
    ix = argmax(gmeans)
    temp["optimal_thresh"] = thresholds[ix]

    plt.plot([0,1], [0,1], linestyle='--')
    plt.plot(fpr, tpr, marker='.', label=f"{model_name} ==> Optimal Threshold={round(thresholds[ix], 2)}")
    plt.scatter(fpr[ix], tpr[ix], marker='o', color='black')
    best_thresholds[i] = temp
# axis labels
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
# show the plot
plt.show()
```

Above graphic shows that Random Forest is the best classification because under the curve are is larger. Also, If I will set threshold 0.01 for random forest I can get better accuracy. Let's do that for all models

```python
for i, model in enumerate(models):
    threshold = best_thresholds[i]['optimal_thresh']
    model_name = str(model).split("(")[0]
    predictions = model.predict_proba(X_test)
    predictions = (predictions[:,1] >= threshold).astype('int')
    acc = calculate_accuracy(y_test.values, predictions)
    print(f"{str(model)} model has {acc} accuracy")
```

As we can compare the accuracies, thresholds increased the accuracy for all models but still not good for production

# Goal 3: Build an unsupervised Machine Learning model and use it to construct a set of events that can be considered as signal events according to your model. Keep in mind that the class labels cannot be used when training the model.

In that goal, I could use Agglomerative Clustering but I prefered to use K-means with cosine distance metric

```python
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
import numpy as np

def kmeans(data:np.array, k:int, iterations:int) -> np.array:
    idx = np.random.choice(len(data), k, replace=False)
    # Selecy random Centroids 
    centroids = data[idx, :]
     
    # calculate the distance between centroids and all data
    distances = cdist(data, centroids ,'cosine')
     
    #Centroid with the minimum Distance
    points = np.array([np.argmin(i) for i in distances])
     
    # Repeating the above steps for a defined number of iterations
    # Repeat the steps until iteration
    for _ in range(iterations): 
        centroids = []
        for idx in range(k):
            
            # Updating Centroids by taking mean of Cluster it belongs to
            temp_cent = data[points==idx].mean(axis=0) 
            centroids.append(temp_cent)
        
        # Updated Centroids 
        centroids = np.vstack(centroids)
         
        distances = cdist(data, centroids ,'cosine')
        points = np.array([np.argmin(i) for i in distances])
         
    return points


pred_labels = kmeans(data_features.values, k=2, iterations=300)

u_labels = np.unique(pred_labels)
plt.subplots(figsize=(25, 12))
for i in u_labels:
    plt.scatter(data_features.values[pred_labels == i , 0] , data_features.values[pred_labels == i , 1] , label = i)
plt.title("K-means Clustering with Cosine Distance Metric")
plt.legend()
plt.show()
```
